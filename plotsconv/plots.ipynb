```julia
using ITensors, ITensorMPS, Plots, LaTeXStrings
using ProgressMeter, Graphs, LinearAlgebra
using Combinatorics, JLD2
using Statistics, Serialization

using Pkg
Pkg.update("JLD2")
```

```julia
s = LinRange(0,5,20)
plot(title = "Average inner product",xlabel=L"\eta")

N=4
means = deserialize("means_N=$(N)_=0.025_maxiter=500_nsamp=100_orthog=true_norm=false.jls")
plot!(s,means,grid=false,fillalpha=.5,label=L"N=%$(N)")


N=8
means = deserialize("means_N=$(N)_=0.025_maxiter=500_nsamp=100_orthog=true_norm=false.jls")
plot!(s,means,grid=false,fillalpha=.5,label=L"N=%$(N)")

N=16
means = deserialize("means_N=$(N)_=0.025_maxiter=500_nsamp=100_orthog=true_norm=false.jls")
plot!(s,means,grid=false,fillalpha=.5,label=L"N=%$(N)")


vline!([1], linestyle=:dash, color=:grey, alpha=0.5, label=L"\eta = 1")
```
