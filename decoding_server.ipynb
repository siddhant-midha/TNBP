## Vacuum

```julia

function plot_results_with_std(n_ar)
    colors = cgrad(:reds, length(n_ar))
    plt = plot()
    for (idx, n) in enumerate(n_ar)
        means_file = "gallagerBP0results/means_all_n=$(n).jld2"
        stds_file = "gallagerBP0results/stds_all_n=$(n).jld2"
        means_data = load(means_file)
        stds_data = load(stds_file)
        ps = means_data["p"]
        means = means_data["means"]
        stds = stds_data["stds"]
        # plot!(plt, ps, means; yerror=stds, marker=:o, color=colors[idx], label="n=$(n)")
        plot!(plt, ps, means; marker=:o, color=colors[idx], label="n=$(n)")

    end
    xlabel!("p")
    title!("Logical error rate")
end

plot_results_with_std([24,48,96])
```
